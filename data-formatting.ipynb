## Data Formatting and Resizing (Don't need to run)

```python
# File path
f_path_sqr = "test_images/train_cfn/"
```

```python
# Rename all images to its shape and enumerated value
files = os.listdir(f_path_sqr)
print(files)
for i, file in enumerate(os.listdir(f_path_sqr)):
    os.rename(f_path_sqr + file, f_path_sqr + "square-" + str(i) + ".png")
```

```python
# Resize all images
## Get file paths
dir_files = os.listdir(f_path_sqr)
all_images = [file for file in dir_files if "square" in file and file.endswith((".png"))]

## Resize images to 200px x 200px
for img in all_images:
    image = Image.open(f_path_sqr + img)
    image.thumbnail(size=(200, 200))
    image.save(f_path_sqr + img, optimize=True)
```
